# Part 1 - Data Preparation and Preprocessing

```python
import string
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display, HTML

# Increase the width of the notebook for displaying DataFrames
display(HTML("<style>.container { width:75% !important; }</style>"))
```

### Load and visualise the dataset

```python
# Low_memory=False allows pandas to get the true datatype for each column by looking through the whole dataset before assigning datatypes.
df = pd.read_csv("data.csv", low_memory=False)
df.head()
```

## 1.2.Normalize and clean the data

### Calculate the avarage salary and store in a new column

```python
df = df.assign(AverageSalary = lambda row: (row.LowestSalary + row.HighestSalary) / 2)
```

### Clean ID and Date column using regex

```python
df["Id"] = df["Id"].replace(to_replace=r'&.*', value='', regex=True)
df["Date"] = df["Date"].replace(to_replace=r'T.*', value='', regex=True)
```

### Change type of related columns

```python
# Set column "Id" to type int64
df["Id"] = pd.to_numeric(df["Id"])

# Set column "Date" to type datetime64[ns]
df["Date"] = pd.to_datetime(df["Date"])
# Data types after data cleaning and conversions
df.dtypes
```

```python
# Visualise the dataset after preparation and preprocessing
df.head()
```

```python
#Save the cleaned and preprocessed dataset for future use if required
df.to_csv("preprocessed_data.csv")
```

# Part 2 - Data Analysis and Interpretation

## Part 2.1 - Analysing the job metadata

### View the number of job postings by classification and subclassification

```python
fig, ax = plt.subplots(1, 2, figsize=(25, 12))

# Extract the necessary information into separate DataFrames
classificationDf = df.groupby(['Classification'], as_index = False)['Id'].count().sort_values(by='Id', ascending=False)
subClassificationDf = df.groupby(['SubClassification'], as_index = False)['Id'].count().sort_values(by='Id', ascending=False)
subClassificationDf = subClassificationDf[:30]

sns.barplot(ax=ax[0], data=classificationDf, x=classificationDf.Id, y=classificationDf.Classification)
sns.barplot(ax=ax[1], data=subClassificationDf, x=subClassificationDf.Id, y=subClassificationDf.SubClassification)

ax[0].set_title("Number of Job Postings by Classification", size=20)
ax[0].set_xlabel("Number of Postings", size=20)
ax[0].set_ylabel("Classification", size=20)

ax[1].set_title("Number of Job Postings by SubClassification", size=20)
ax[1].set_xlabel("Number of Postings", size=20)
ax[1].set_ylabel("SubClassification", size=20)

plt.subplots_adjust(wspace=0.5)
```

### View market share by location

```python
locationDf = df[df['Location'].isna()==False]
locationDf = locationDf.groupby('Location').count()
locationDf['Location'] = locationDf.index
locationDf['Count'] = locationDf["Id"]

#Sort by descending to have the top 5 Cities at the top
locationDf = locationDf.sort_values(by="Count", ascending=False)
#Copy the top 5 countries
df2 = locationDf[:6].copy()
#Add a new row that will be all of the countries under the top 5, rename to
#others and set as the sum of all of their counts
new_row = pd.DataFrame(data = {'Location' : ['others'],
                              'Count' : [locationDf['Count'][6:].sum()]
                              })
#Overwrite the dataframe with the concatenation of these two
locationDf = pd.concat([df2, new_row])
locationDf.head()

colors = sns.color_palette('pastel')
fig = plt.figure(figsize=(10, 10))
plt.pie(data=locationDf, x=locationDf.Count, labels= locationDf.Location, labeldistance=None, colors=colors, autopct='%0.1f%%', 
         wedgeprops={'edgecolor':'black','linewidth':3}, rotatelabels=True, pctdistance=0.8)
plt.legend(prop={"size":12})
plt.title("Market share of locations",size=20);
```

### View job postings by average salary

```python
#Group by average salary and count the numbers
averageSalaryDf = df.groupby('AverageSalary').count()
averageSalaryDf['AverageSalary'] = (averageSalaryDf.index*1000).astype(float).map('{:,.0f}'.format)
averageSalaryDf['Count'] = averageSalaryDf["Id"]

labels = averageSalaryDf["AverageSalary"]

colors = sns.color_palette('pastel')
fig = plt.figure(figsize=(10, 10))
plt.pie(data=averageSalaryDf, x=averageSalaryDf.Id, labels=labels, colors=colors, autopct='%0.2f%%',
       wedgeprops={'edgecolor':'black','linewidth':3}, textprops={'fontsize': 12}, pctdistance=0.8)

plt.title("Job Posts by Average Salary", size=20);
```

### View the distribution of job types

```python
JobType = df.groupby('JobType', as_index=False).count()
JobType['labels'] = JobType['JobType']

colors = sns.color_palette('pastel')
fig = plt.figure(figsize=(10, 10))
plt.pie(data=JobType, x=JobType.Id, labels=JobType.labels, colors=colors, autopct='%0.2f%%',
       wedgeprops={'edgecolor':'black','linewidth':3}, textprops={'fontsize': 12}, labeldistance=None)
plt.title("Job Type Distribution", size=20)
plt.legend(prop={"size":12});
```

## Sydney

### Area and Classification

```python
Sydney_Df = df[df['Location'].isna()==False]
Sydney_Df = Sydney_Df[Sydney_Df['Location'] == "Sydney"]

Sydney_AreaDf = Sydney_Df.groupby('Area').count()
Sydney_AreaDf['Area'] = Sydney_AreaDf.index
Sydney_AreaDf['Count'] = Sydney_AreaDf['Id']

Sydney_ClassDf = Sydney_Df.groupby('Classification').count()
Sydney_ClassDf['Classification'] = Sydney_ClassDf.index
Sydney_ClassDf['Count'] = Sydney_ClassDf['Id']
```

### Job posting dates

```python
FMFH = 0
FMSH = 0
SMFH = 0
SMSH = 0

for i in range(len(Sydney_Df['Date'])):
    if Sydney_Df['Date'].iloc[i].month == 10 and Sydney_Df['Date'].iloc[i].day <= 15:
        FMFH += 1 #FIRST MONTH FIRST HALF
    elif Sydney_Df['Date'].iloc[i].month == 10 and Sydney_Df['Date'].iloc[i].day > 15 \
    and Sydney_Df['Date'].iloc[i].day <= 31:
        FMSH += 1 #SECOND MONTH SECOND HALF
    elif Sydney_Df['Date'].iloc[i].month == 11 and Sydney_Df['Date'].iloc[i].day <= 15:
        SMFH += 1
    elif Sydney_Df['Date'].iloc[i].month == 11 and Sydney_Df['Date'].iloc[i].day > 15 \
        and Sydney_Df['Date'].iloc[i].day <= 31:
        SMSH += 1

PostingData = {'PostDate': ['1st to 15th - 10 - 2018',
                                '16th to 31st - 10 - 2018',
                                '1st to 15th - 11 - 2018',
                                '16th to 30th - 11 - 2018'],
                   'PostFrequency': [FMFH, FMSH, SMFH, SMSH]}

Sydney_PostingData_Df = pd.DataFrame(PostingData)
Sydney_PostingData_Df = Sydney_PostingData_Df.sort_values(by="PostFrequency", ascending=False)
Sydney_PostingData_Df = Sydney_PostingData_Df[:3]
```

```python
#Sort by descending to have the top 5 AREAS at the top
Sydney_AreaDf = Sydney_AreaDf.sort_values(by="Count", ascending=False)
#Copy the top 5 countries
df2 = Sydney_AreaDf[:6].copy()
#Add a new row that will be all of the AREAS under the top 5, rename to
#others and set as the sum of all of their counts
new_row = pd.DataFrame(data = {'Area' : ['others'],
                              'Count' : [Sydney_AreaDf['Count'][6:].sum()]
                              })
#Overwrite the dataframe with the concatenation of these two
Sydney_AreaDf = pd.concat([df2, new_row])

#Sort by descending to have the top 5 CLASSIFICATIONS at the top
Sydney_ClassDf = Sydney_ClassDf.sort_values(by="Count", ascending=False)
#Copy the top 5 countries
df2 = Sydney_ClassDf[:14].copy()
#Add a new row that will be all of the CLASSIFICATIONS under the top 5, rename to
#others and set as the sum of all of their counts
new_row = pd.DataFrame(data = {'Classification' : ['others'],
                              'Count' : [Sydney_ClassDf['Count'][14:].sum()]
                              })
#Overwrite the dataframe with the concatenation of these two
Sydney_ClassDf = pd.concat([df2, new_row])
# Sydney_AreaDf.head(5)
```

```python
#Group by average salary and count the numbers
SydneyAverageSalaryDf = Sydney_Df.groupby('AverageSalary').count()
SydneyAverageSalaryDf['AverageSalary'] = (SydneyAverageSalaryDf.index*1000).astype(float).map('{:,.0f}'.format)
SydneyAverageSalaryDf['Count'] = SydneyAverageSalaryDf["Id"]
SydneyAverageSalaryDf

labels = SydneyAverageSalaryDf["AverageSalary"]
```

```python
colors = sns.color_palette('pastel')[0:7]
fig = plt.figure(figsize=(22, 20))
plt.subplot(222)
plt.pie(data=Sydney_AreaDf, x=Sydney_AreaDf.Count, labels=Sydney_AreaDf.Area, colors=colors, autopct='%0.1f%%',
       wedgeprops={'edgecolor':'black','linewidth':3}, textprops={'fontsize': 16}, pctdistance=0.8, labeldistance=None)
plt.title("Sydney Market share of Areas", size=30)
plt.legend()

plt.subplot(221)
plt.pie(data=Sydney_ClassDf, x=Sydney_ClassDf.Count, labels=Sydney_ClassDf.Classification, colors=colors, autopct='%0.1f%%',
       wedgeprops={'edgecolor':'black','linewidth':3}, textprops={'fontsize': 16}, pctdistance=0.85)
plt.title("Sydney Market share of Job Classifications", size=30)

plt.subplot(223)
plt.pie(data=SydneyAverageSalaryDf, x=SydneyAverageSalaryDf.Id, labels=labels, colors=colors, autopct='%0.2f%%',
       wedgeprops={'edgecolor':'black','linewidth':3}, textprops={'fontsize': 16}, pctdistance=0.8)

plt.title("Sydney Job Posts by Average Salary", size=30);

plt.subplot(224)
plt.pie(data=Sydney_PostingData_Df, x=Sydney_PostingData_Df['PostFrequency'], labels=Sydney_PostingData_Df['PostDate'], colors=colors, autopct='%0.1f%%',
       wedgeprops={'edgecolor':'black','linewidth':3}, textprops={'fontsize': 16}, pctdistance=0.8, labeldistance=None)
plt.title("Sydney Job Posting Dates", size=30)
plt.legend()

plt.subplots_adjust(wspace = 0.315)
```
