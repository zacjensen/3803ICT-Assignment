## Which sectors keep the highest market share?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
df = pd.read_csv("data.csv", low_memory=False)
print(f"The dataset has {df.shape[0]} rows and {df.shape[1]} columns.")
df['Date'] = pd.to_datetime(df['Date']) # The data is changed to datetime var type
df = df.assign(AverageSalary = lambda row: (row.LowestSalary + row.HighestSalary) / 2)
df['AverageSalary'] = (df.index*1000).astype(float).map('{:,.0f}'.format)
df.head()
```

```python
by_class = df.groupby(['Classification'], as_index = False)['Id'].count().sort_values(by='Id', ascending=False)
fig = plt.figure(figsize=(10, 10))
sns.barplot(data=by_class, x=by_class.Id, y=by_class.Classification)
plt.title("Classifications of Jobs and Number of Postings",size=20)
plt.xlabel("Number of Postings",size=20)
plt.ylabel("Classification",size=20)
```

```python
by_Subclass = df.groupby(['SubClassification'], as_index = False)['Id'].count().sort_values(by='Id', ascending=False)
by_Subclass = by_Subclass[:30]
fig = plt.figure(figsize=(10, 10))
sns.barplot(data=by_Subclass, x=by_Subclass.Id, y=by_Subclass.SubClassification)
plt.title("Subclassifications of Jobs and Number of Postings",size=20)
plt.xlabel("Number of Postings",size=20)
plt.ylabel("SubClassification",size=20)
```

```python
SubClassification1Df = df[df['Classification'] == by_class['Classification'].iloc[0]].groupby('SubClassification', as_index=False).count().sort_values(by='Id', ascending=False)
SubClassification3Df = df[df['Classification'] == by_class['Classification'].iloc[2]].groupby('SubClassification', as_index=False).count().sort_values(by='Id', ascending=False)
SubClassification4Df = df[df['Classification'] == by_class['Classification'].iloc[3]].groupby('SubClassification', as_index=False).count().sort_values(by='Id', ascending=False)

df2 = SubClassification1Df[:5].copy()
#Add a new row that will be all of the AREAS under the top 5, rename to
#others and set as the sum of all of their counts
new_row = pd.DataFrame(data = {'SubClassification' : ['others'],
                              'Id' : [SubClassification1Df['Id'][5:].sum()]
                              })
print(new_row)
#Overwrite the dataframe with the concatenation of these two
SubClassification1Df = pd.concat([df2, new_row])

SubClassification1Df.head(10)

# SubClassification1Df = SubClassification1Df[:5]



print("Subclassification:", by_class['Classification'].iloc[0])
colors = sns.color_palette('pastel')
fig = plt.figure(figsize=(15, 10))
plt.subplot(121)
plt.pie(data=SubClassification1Df, x=SubClassification1Df.Id, labels= SubClassification1Df.SubClassification, labeldistance=None, colors=colors, autopct='%0.1f%%', 
         wedgeprops={'edgecolor':'black','linewidth':3}, rotatelabels=None, pctdistance=0.8, explode=[0, 0, 0, 0, 0, 0.25])
plt.legend()
plt.title("ICT Sub-Sector Spotlights", size=20)

plt.subplot(122)
SubClassification1Df = SubClassification1Df[:5]
plt.pie(data=SubClassification1Df, x=SubClassification1Df.Id, labels= SubClassification1Df.SubClassification, labeldistance=None, colors=colors, autopct='%0.1f%%', 
         wedgeprops={'edgecolor':'black','linewidth':3}, rotatelabels=None, pctdistance=0.8, explode=[0, 0, 0, 0, 0])
plt.legend(loc=3)
plt.title("Top 5 ICT Sub-Sector Spotlights", size=20)

SubClassification1Df.head() 
```
